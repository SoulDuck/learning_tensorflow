# Package Load

```python
import preprocessing
import tensorflow as tf
import numpy as np
import random
import sys
import utils
import pickle
from IPython.display import Image
# preprocessing : 이미지 파일로 존재하는 데이터를 numpy와 onehotencoding 라벨로 바꿔주는 모듈
# tensorflow : 텐서플로우 , 일반적으로 tf로 줄여서 사용
# numpy : fundamental package for scientific computing with Python
# pickle : python 의 데이터 형식을 그대로 저장하기 위한 파이썬 기본 패키지
```

# Image Loading

```python
# 이미지 파일이 있는 폴더를 parameter 로 입력
train_imgs , train_labs =preprocessing.get_imagenet_data("/home/mediwhale-3/imagenet",1000)


def next_batch(imgs , labs , batch_size):
    indices=random.sample(range(len(imgs)) , batch_size)
    batch_xs=imgs[indices]
    batch_ys=labs[indices]
    return batch_xs , batch_ys

# preprocessing 모듈에서 train_imgs , train_labs , test_imgs , test_labs 를 불러 온다.
# fully connected 모델을 위해 32x32x3 크기의 이미지를 3072 으로 reshape 한다.
# next_batch 함수는 batch_size 수 만큼 random 으로 추출하여 train batch 를 생성하는 함수
```

# Set Placeholder And Variables

```python
height = 224
width = 224
color_ch=3
n_classes = 1000
learning_rate=0.001
max_iter=10000
check_point=100
x_ =tf.placeholder(tf.float32, [ None , height , width , color_ch ])
y_ =tf.placeholder( tf.int32 , [ None , n_classes ])
phase_train = tf.placeholder(dtype=tf.bool,name="phase_train")
# Placeholder 는 차후에 입력할 값
# Variables 는 weight 로 학습시 변하는 값
```

# Network Modeling

```python
def init_weights(shape,name):
    return tf.get_variable(name , shape , initializer=tf.contrib.layers.xavier_initializer())

weights = {
    "w1" : init_weights([7,7,3,64],"w1"),
    "r1_1" : init_weights([3,3,64,64],"r1_1"),
    "r1_2" : init_weights([3,3,64,64],"r1_2"),
    "r2_1" : init_weights([3,3,64,64],"r2_1"),
    "r2_2" : init_weights([3,3,64,64],"r2_2"),
    "r3_1" : init_weights([3,3,64,64],"r3_1"),
    "r3_2" : init_weights([3,3,64,64],"r3_2"),
    "r4_1" : init_weights([3,3,64,128],"r4_1"),
    "r4_2" : init_weights([3,3,128,128],"r4_2"),
    "r4_s" : init_weights([1,1,64,128],"r4_s"),
    "r5_1" : init_weights([3,3,128,128],"r5_1"),
    "r5_2" : init_weights([3,3,128,128],"r5_2"),
    "r6_1" : init_weights([3,3,128,128],"r6_1"),
    "r6_2" : init_weights([3,3,128,128],"r6_2"),
    "r7_1" : init_weights([3,3,128,128],"r7_1"),
    "r7_2" : init_weights([3,3,128,128],"r7_2"),
    "r8_1" : init_weights([3,3,128,256],"r8_1"),
    "r8_2" : init_weights([3,3,256,256],"r8_2"),
    "r8_s" : init_weights([1,1,128,256],"r8_s"),
    "r9_1" : init_weights([3,3,256,256],"r9_1"),
    "r9_2" : init_weights([3,3,256,256],"r9_2"),
    "r10_1" : init_weights([3,3,256,256],"r10_1"),
    "r10_2" : init_weights([3,3,256,256],"r10_2"),
    "r11_1" : init_weights([3,3,256,256],"r11_1"),
    "r11_2" : init_weights([3,3,256,256],"r11_2"),
    "r12_1" : init_weights([3,3,256,256],"r12_1"),
    "r12_2" : init_weights([3,3,256,256],"r12_2"),
    "r13_1" : init_weights([3,3,256,256],"r13_1"),
    "r13_2" : init_weights([3,3,256,256],"r13_2"),
    "r14_1" : init_weights([3,3,256,512],"r14_1"),
    "r14_2" : init_weights([3,3,512,512],"r14_2"),
    "r14_s" : init_weights([1,1,256,512],"r14_s"),
    "r15_1" : init_weights([3,3,512,512],"r15_1"),
    "r15_2" : init_weights([3,3,512,512],"r15_2"),
    "r16_1" : init_weights([3,3,512,512],"r16_1"),
    "r16_2" : init_weights([3,3,512,512],"r16_2"),
    
    "w_f" : init_weights([512,1000],"w_f"),
    "b_f" : init_weights([1000],"b_f")
}

layer = tf.nn.conv2d(x_, weights["w1"], strides=[1, 2, 2, 1], padding='SAME')
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

layer = tf.nn.max_pool(layer, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

residual = tf.nn.conv2d(layer, weights["r1_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r1_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r2_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r2_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r3_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r3_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r4_1"], strides=[1, 2, 2, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r4_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.nn.conv2d(layer, weights["r4_s"], strides=[1, 2, 2, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r5_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r5_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r6_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r6_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r7_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r7_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r8_1"], strides=[1, 2, 2, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r8_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.nn.conv2d(layer, weights["r8_s"], strides=[1, 2, 2, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r9_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r9_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r10_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r10_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r11_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r11_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r12_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r12_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r13_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r13_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r14_1"], strides=[1, 2, 2, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r14_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.nn.conv2d(layer, weights["r14_s"], strides=[1, 2, 2, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r15_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r15_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

residual = tf.nn.conv2d(layer, weights["r16_1"], strides=[1, 1, 1, 1], padding='SAME')
residual = tf.contrib.layers.batch_norm(residual, center=True, scale=True, is_training=phase_train)
residual = tf.nn.relu(residual)
residual = tf.nn.conv2d(residual, weights["r16_2"], strides=[1, 1, 1, 1], padding='SAME')

layer = tf.add(layer,residual)
layer = tf.contrib.layers.batch_norm(layer, center=True, scale=True, is_training=phase_train)
layer = tf.nn.relu(layer)

layer = tf.nn.avg_pool(layer, ksize=[1, 7, 7, 1], strides=[1, 7, 7, 1], padding='SAME')
layer = tf.reshape(layer, [-1,512])

layer = tf.nn.softmax(tf.add(tf.matmul(layer,weights['w_f']), weights['b_f']))



end_layer = layer
```

# Define Functions

```python
pred=tf.nn.softmax(end_layer)
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits= pred , labels=y_) , name='cost')
train=tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost,name='train')
correct_pred = tf.equal(tf.argmax(pred, 1), tf.argmax(y_, 1), name='correct_pred')
accuracy = tf.reduce_mean(tf.cast(correct_pred, dtype=tf.float32), name='accuracy')

# softmax : 결과값을 각 class 에 해당하는 확률값을로 리턴
# softmax_cross_entropy_with_logits : 실제 정답인 onehot vector 와 예측값 pred 를 차이를 cross_entropy 로 계산
# tf.train.GradientDescentOptimizer : cost 가 최소와 되도록 weight를 조정하는 함수
# accuracy : 실제 값과 예측값의 일치률
```

# Training

```python
sess=tf.Session()
init=tf.global_variables_initializer()
sess.run(init)
saver = tf.train.Saver()

batch_iteration = 100    
training_epochs = 2000

train_cost_list = []
test_cost_list = []
train_acc_list = []
test_acc_list = []

# Tensorflow 학습을 위한 session 정의
# saver : 학습한 weight 와 graph 를 저장

# sess.run 을 통해 함수를 실행 , feed_dict 는 placeholder 에 들어갈 값

for epoch in range(training_epochs):
    
    train_avg_cost = 0.
    train_avg_acc = 0.
    test_avg_cost = 0.
    test_avg_acc = 0.
    
    for batch in range(batch_iteration) :
        batch_xs , batch_ys = next_batch(train_imgs , train_labs , 20)
        
        sess.run(train, feed_dict={x_:batch_xs , y_ : batch_ys,phase_train:True})
        train_avg_cost += sess.run(cost, feed_dict={x_:batch_xs , y_ : batch_ys,phase_train:False})
        train_avg_acc += sess.run(accuracy, feed_dict={x_:batch_xs , y_ : batch_ys,phase_train:False})
        
    train_avg_cost = train_avg_cost / batch_iteration
    train_avg_acc = train_avg_acc / batch_iteration
    
#     test_avg_cost = sess.run(cost, feed_dict={x_: test_imgs, y_: test_labs})
#     test_avg_acc = sess.run(accuracy, feed_dict={x_: test_imgs, y_: test_labs})
    
    
    print "##################################"
    print ("Epoch: %03d/%03d train cost: %.4f" % (epoch, training_epochs, train_avg_cost))
    print ("Epoch: %03d/%03d train acc: %.4f" % (epoch, training_epochs, train_avg_acc))
#     print ("Epoch: %03d/%03d test cost: %.4f" % (epoch, training_epochs, test_avg_cost))
#     print ("Epoch: %03d/%03d test acc: %.4f" % (epoch, training_epochs, test_avg_acc))    
    
#     train_cost_list.append(train_avg_cost)
#     test_cost_list.append(test_avg_cost)
#     train_acc_list.append(train_avg_acc)
#     test_acc_list.append(test_avg_acc)
   
    
    
# saver.save(sess, "model/cifar_deep_convolution.ckpt")

# with open('cost_acc/cifar_deep_convolution_train_cost', 'wb') as fp:
#     pickle.dump(train_cost_list, fp)
# with open('cost_acc/cifar_deep_convolution_test_cost', 'wb') as fp:
#     pickle.dump(test_cost_list, fp)
# with open('cost_acc/cifar_deep_convolution_train_acc', 'wb') as fp:
#     pickle.dump(train_acc_list, fp)
# with open('cost_acc/cifar_deep_convolution_test_acc', 'wb') as fp:
#     pickle.dump(test_acc_list, fp)
```
